### 引入函式庫

```python
# My code 
import requests
from bs4 import BeautifulSoup
```

### 以requests.get得到網址內容，並使用BeautifulSoup解析結構後，印出該結構的Text形式

```python
res  = requests.get('https://github.com/new')
soup = BeautifulSoup(res.text)

for i in soup.text.split('\n'):
    if i == '':
        continue
    else:
        print(i)
```

## 利用請求中帶入 cookie 的方式登入網頁

```python
res = requests.get('https://github.com/new',verify = False, cookies={'user_session': 'FqeiwApghXGQRlEq7Cy1NVk0rSO5JtzbLOl-PYq-mP-oioKH'})
soup = BeautifulSoup(res.text)

for i in soup.text.split('\n'):
    if i == '' or i == '↵':
        continue
    else:
        print(i)
```
