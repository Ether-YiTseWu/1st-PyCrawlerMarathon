### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異
- json.loads(r.text) 一個裡，[ ] 中便是一個陣列。因此json.loads(r.text)中有很多陣列
- r.text只有一個陣列，便是r.text自己

```python
# My code
import requests

r = requests.get('https://api.github.com/events')
r.text[:1500]
```

```python
# My code
import json

jsonData = json.loads(r.text)
jsonData[:1]
```
