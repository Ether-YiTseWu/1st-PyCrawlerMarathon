## 作業目標

* 根據範例提供的 API ，完成以下問題：

    1. 取出知乎問題發問時間
    2. 取出第一筆與最後一筆回答的時間


```python
# My code
import requests
import json

headers = {'user-agent': 'my-app/0.0.1'}
r = requests.get('https://www.zhihu.com/api/v4/questions/55493026/answers',headers=headers)
response = r.text

jsonData = json.loads(response)
jsonData
```

### 1. 取出知乎問題發問時間

```python
# 1. 取出知乎問題發問時間

# My code
from datetime import datetime

for i in jsonData['data']:
    # 時間需要經過轉換
    time = datetime.utcfromtimestamp(i['created_time'])
    print(time)
```

### 2. 取出第一筆與最後一筆回答的時間

```python
# 2. 取出第一筆與最後一筆回答的時間

# My code
time = []
for i in jsonData['data']:
    time.append(datetime.utcfromtimestamp(i['created_time']))
    
print("第一筆回答的時間 :",time[0])
print("最後筆回答的時間 :",time[len(time)-1])
```
